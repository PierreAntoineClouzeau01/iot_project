```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Crop_recommendation.csv')
```

```python
len(df.label.value_counts())
```

```python
df_rice = df[df.label == 'rice']
```

```python
plt.figure(figsize=(10,15))

for i, k in enumerate(df.iloc[:, :-1]):
    plt.subplot(4,2, i+1)
    

    sns.kdeplot(df[df.label == 'rice'][k])
    plt.title(k)
```

```python
for i, k in enumerate(df_rice.iloc[:, :-1]) :
    plt.subplot(4, 2, i+1)
    plt.plot(df_rice.index, df_rice[k])
```
